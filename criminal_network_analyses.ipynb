```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Dictionary to store phases and graphs
phases = {}
G = {}

# List of players under investigation
list_of_players = [1, 3, 83, 86, 85, 6, 11, 88, 106, 89, 84, 5, 8, 76, 77, 87, 82, 96, 12, 17, 80, 33, 16]
players = ["n" + str(player) for player in list_of_players]  # Convert to node names

# Store results for all phases
centrality_results = []

# Loop through each phase
for i in range(1, 12):
    var_name = f"phase{i}"
    file_name = f"https://raw.githubusercontent.com/ragini30/Networks-Homework/main/{var_name}.csv"

    # Load the data and preprocess it
    phases[i] = pd.read_csv(file_name, index_col=["players"])
    phases[i].columns = "n" + phases[i].columns
    phases[i].index = phases[i].columns
    phases[i][phases[i] > 0] = 1  # Binarize the adjacency matrix

    # Create NetworkX graph
    G[i] = nx.from_pandas_adjacency(phases[i])
    G[i].name = var_name

    # Compute centrality measures
    degree_centrality = nx.degree_centrality(G[i])
    betweenness_centrality = nx.betweenness_centrality(G[i], normalized=True)
    try:
        eigenvector_centrality = nx.eigenvector_centrality(G[i])
    except nx.PowerIterationFailedConvergence:
        eigenvector_centrality = {node: 0 for node in G[i].nodes()}

    # Filter for the 23 players under investigation
    player_degree = {p: degree_centrality.get(p, 0) for p in players}
    player_betweenness = {p: betweenness_centrality.get(p, 0) for p in players}
    player_eigenvector = {p: eigenvector_centrality.get(p, 0) for p in players}

    # Find top 5 players for each centrality measure
    top5_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
    top5_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
    top5_eigenvector = sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)[:5]

    # Print top 5 players for each centrality measure
    print(f"\nPhase {i} - Top 5 Players:")
    print("Degree Centrality:")
    for player, score in top5_degree:
        print(f"{player}: {score:.4f}")
    print("Betweenness Centrality:")
    for player, score in top5_betweenness:
        print(f"{player}: {score:.4f}")
    print("Eigenvector Centrality:")
    for player, score in top5_eigenvector:
        print(f"{player}: {score:.4f}")

    # Store results in a structured way
    for p in G[i].nodes():
        centrality_results.append({
            "Phase": i,
            "Player": p,
            "Degree Centrality": degree_centrality.get(p, 0),
            "Betweenness Centrality": betweenness_centrality.get(p, 0),
            "Eigenvector Centrality": eigenvector_centrality.get(p, 0),
            "In Top 5 (Degree)": p in dict(top5_degree),
            "In Top 5 (Betweenness)": p in dict(top5_betweenness),
            "In Top 5 (Eigenvector)": p in dict(top5_eigenvector),
            "Investigated Player": p in players
        })

    # Visualize the network with node size based on degree centrality
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G[i], seed=42)
    node_sizes = [degree_centrality[n] * 2000 for n in G[i].nodes()]
    nx.draw(G[i], pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=node_sizes, font_size=10)
    plt.title(f"Network for {var_name}")
    plt.show()

# Convert results to a DataFrame and save as CSV
centrality_df = pd.DataFrame(centrality_results)
centrality_df.to_csv("centrality_results.csv", index=False)
print("Centrality results saved to 'centrality_results.csv'")
```

```python
import pandas as pd
import networkx as nx

# Dictionary to store phases and graphs
phases = {}
G_binary = {}
G_weighted = {}

# List of players under investigation
list_of_players = [1, 3, 83, 86, 85, 6, 11, 88, 106, 89, 84, 5, 8, 76, 77, 87, 82, 96, 12, 17, 80, 33, 16]
players = ["n" + str(player) for player in list_of_players]  # Convert to node names

# Store results for all phases
centrality_results = []

# Loop through each phase
for i in range(1, 12):
    var_name = f"phase{i}"
    file_name = f"https://raw.githubusercontent.com/ragini30/Networks-Homework/main/{var_name}.csv"

    # Load the data and preprocess it
    phases[i] = pd.read_csv(file_name, index_col=["players"])
    phases[i].columns = "n" + phases[i].columns
    phases[i].index = phases[i].columns

    # Create binary adjacency matrix
    binary_matrix = phases[i].copy()
    binary_matrix[binary_matrix > 0] = 1

    # Create directed graphs for binary and weighted matrices
    G_binary[i] = nx.from_pandas_adjacency(binary_matrix, create_using=nx.DiGraph())
    G_weighted[i] = nx.from_pandas_adjacency(phases[i], create_using=nx.DiGraph())

    # Compute centrality measures for binary graph
    degree_centrality_binary = nx.degree_centrality(G_binary[i])
    betweenness_centrality_binary = nx.betweenness_centrality(G_binary[i], normalized=True)
    try:
        eigenvector_centrality_binary = nx.eigenvector_centrality(G_binary[i])
    except nx.PowerIterationFailedConvergence:
        eigenvector_centrality_binary = {node: 0 for node in G_binary[i].nodes()}
    hubs_binary, authorities_binary = nx.hits(G_binary[i], max_iter=1000000)

    # Compute centrality measures for weighted graph
    degree_centrality_weighted = nx.degree_centrality(G_weighted[i])
    betweenness_centrality_weighted = nx.betweenness_centrality(G_weighted[i], normalized=True, weight='weight')
    try:
        eigenvector_centrality_weighted = nx.eigenvector_centrality(G_weighted[i], weight='weight')
    except nx.PowerIterationFailedConvergence:
        eigenvector_centrality_weighted = {node: 0 for node in G_weighted[i].nodes()}
    hubs_weighted, authorities_weighted = nx.hits(G_weighted[i], max_iter=1000000)

    # Store results in a structured way
    for p in players:
        centrality_results.append({
            "Phase": i,
            "Player": p,
            "Degree Centrality (Binary)": degree_centrality_binary.get(p, 0),
            "Degree Centrality (Weighted)": degree_centrality_weighted.get(p, 0),
            "Betweenness Centrality (Binary)": betweenness_centrality_binary.get(p, 0),
            "Betweenness Centrality (Weighted)": betweenness_centrality_weighted.get(p, 0),
            "Eigenvector Centrality (Binary)": eigenvector_centrality_binary.get(p, 0),
            "Eigenvector Centrality (Weighted)": eigenvector_centrality_weighted.get(p, 0),
            "Hub Score (Binary)": hubs_binary.get(p, 0),
            "Hub Score (Weighted)": hubs_weighted.get(p, 0),
            "Authority Score (Binary)": authorities_binary.get(p, 0),
            "Authority Score (Weighted)": authorities_weighted.get(p, 0)
        })

# Convert results to a DataFrame and save as CSV
centrality_df = pd.DataFrame(centrality_results)
centrality_df.to_csv("centrality_comparison_results.csv", index=False)
print("Centrality comparison results saved to 'centrality_comparison_results.csv'")

from google.colab import files

# Convert results to a DataFrame and save as CSV
centrality_df = pd.DataFrame(centrality_results)
centrality_df.to_csv("centrality_comparison_results.csv", index=False)

# Download the file
files.download("centrality_comparison_results.csv")
```
